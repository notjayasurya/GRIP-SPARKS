```python
# Decision tree classification using the iris dataset and representing it visually.
# Importing the needed libraries for the classification
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("C:/Users/tkjay/Desktop/iris.csv")
```

```python
iris.head()
```

```python
# Finding the relations between the columns (just for cross referance)
sns.pairplot(iris, hue = 'species')
```

```python
# Importing LabelEncoder for label encoding the ['species'] column
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
iris["species"] = LE.fit_transform(iris["species"])
```

```python
iris.head()
```

```python
#Making the data ready for going through training and spliting.
x = iris[["sepal_length","sepal_width","petal_length","petal_width"]]
y = iris["species"]
```

```python
x.head()
```

```python
y.head()
```

```python
#importing and training the data.
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
(x_train,x_text,y_train,y_test)= train_test_split(x,y,train_size = 0.7, random_state = 1)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
#Representing the decision tree graphically.
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
decisiontree = DecisionTreeClassifier().fit(x, y)
# Prepare a plot figure with set size.
plt.figure(figsize = (20,10))
# Plot the decision tree, showing the decisive values and the improvements in Gini impurity along the way.
plot_tree(decisiontree, rounded = True, filled=True)
# Display the tree plot figure.
plt.show()
```
